# outlier for x

```python
q1=np.percentile(df['x'],25, interpolation="midpoint") #outlier for x
q2=np.percentile(df['x'],50, interpolation="midpoint")
q3=np.percentile(df['x'],75, interpolation="midpoint")
print(q1)
print(q2)
print(q3)
```

```python
df['x'].median()
```

```python
IQR= q3-q1
IQR
```

```python
lower_limit= q1 - 1.5*IQR
upper_limit= q3 + 1.5*IQR

print(lower_limit)
print(upper_limit)
```

```python
outlier=[]
for x in df["x"]:
    if (x>upper_limit) or (x<lower_limit) :
        outlier.append(x)
```

```python
outlier
```

```python
df.drop(df[(df["x"]>upper_limit)|(df["x"]<lower_limit)].index, inplace= True)
```

```python
df.describe()
```

```python
plt.boxplot(df['x'])
plt.title("box plot of x")
plt.show()
```
